```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import random
import time

# Set up Chrome with your profile
chrome_driver_path = "/Users/fadil/Downloads/chromedriver-mac-arm642/chromedriver"
chrome_options = Options()
chrome_options.add_argument("--user-data-dir=/Users/fadil/Library/Application Support/Google/Chrome")  
chrome_options.add_argument("--profile-directory=Default")  
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

def human_delay(min_time=2, max_time=5):
    """Waits for a random time between actions to mimic human behavior."""
    delay = random.uniform(min_time, max_time)
    time.sleep(delay)

# Scroll to load more posts
def human_scroll(driver, times=5):
    """Scrolls randomly like a human user to load more posts."""
    for _ in range(times):
        driver.execute_script("window.scrollBy(0, 800);")
        time.sleep(2)





# Navigate to Facebook group/page
driver.execute_script("window.open('https://www.facebook.com/bebephone.online','_blank');")  # Replace with actual URL
human_delay(3, 6) 

# Switch to the new tab
driver.switch_to.window(driver.window_handles[-1])
print(" Switched to new tab.")

human_scroll(driver, 5)
# Keywords to search for in posts
keywords = ["power bank", "‡∏û‡∏≤‡∏ß‡πÄ‡∏ß‡∏≠‡∏£‡πå‡πÅ‡∏ö‡∏á‡∏Ñ‡πå", "HAPPY VALENTINE DAY!", "‡∏û‡∏≤‡∏ß‡πÄ‡∏ß‡∏≠‡∏£‡πå‡πÅ‡∏ö‡∏á‡∏Ñ‡πå‡∏ï‡∏±‡∏ß‡∏à‡∏ö"]

# Find posts
post_containers = driver.find_elements(By.XPATH, "//div[contains(@role,'article')]")
print(f" Found {len(post_containers)} posts on the page!")

# Create folder for images
output_folder = "facebook_powerbank_images1"
os.makedirs(output_folder, exist_ok=True)

found_any = False  # Track if any post was found

for index, post in enumerate(post_containers):
    try:
        post_text = post.text.lower()
        
        # Extract post link (if available)
        try:
            post_link = post.find_element(By.XPATH, ".//a[contains(@href, '/posts/')]").get_attribute("href")
            post_id = post_link.split("posts/")[-1].split("/")[0]  # Extract post ID from link
        except:
            post_link = "No_Link_Found"
            post_id = f"post_{index + 1}_{int(time.time())}"  # Fallback ID if no link found
        
        print(f"\n**Post {index + 1}:** \n{post_text[:200]}...")  # Print first 200 chars of the post
        print(f"Post Link: {post_link}")

        if any(keyword in post_text for keyword in keywords):
            found_any = True
            print("**Keyword Detected! Extracting post images...**")

            # Find image elements in the post (excluding profile pictures & stickers)
            images = post.find_elements(By.XPATH, ".//img")

            for idx, img in enumerate(images):
                img_url = img.get_attribute("src")

                # Filter only relevant post images (skip emojis, small icons, etc.)
                if img_url and "profile" not in img_url and "emoji" not in img_url and "sticker" not in img_url:
                    print(f"Image {idx + 1} URL: {img_url}")

                    # Save all images from the same post with the same prefix
                    img_filename = os.path.join(output_folder, f"{post_id}_image_{idx + 1}.jpg")

                    # Download image
                    response = requests.get(img_url, stream=True)
                    if response.status_code == 200:
                        with open(img_filename, "wb") as file:
                            for chunk in response.iter_content(1024):
                                file.write(chunk)
                        print(f"üì∏ Image saved: {img_filename} (Linked to: {post_link})")
                    else:
                        print(f"Failed to download image: {img_url}")

    except Exception as e:
        print(f"‚ö†Ô∏è Error processing post: {e}")

if not found_any:
    print("\nNo matching posts found!")

# Close the driver after scraping
driver.quit()


```
